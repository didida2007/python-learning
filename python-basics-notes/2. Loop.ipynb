## 流程
* 管理计算机代码的执行顺序

#### 流程分类
1. 顺序流程：自上而下的执行结构（默认）
2. 选择流程：根据在某一步的判断，选择执行相对应的逻辑\
    i. 单选：if\
    ii. 二选一：if ... else\
    iii. 多选一： if ... elif ... else
3. 循环流程：在一定条件下，一直重复执行某段代码的逻辑\
    while\
    for ... in ...

## 循环流程

```python
index = 1
while index <= 100:
    print("loop %d" %index)
    index += 1
```

```python
num1 = 1
while num1 <= 9:
    num2 = 1
    while num2 <= 10 - num1:
        print("%d * %d = %d" %(num1, num2, num1 * num2), end = "   ")   # 句子结尾不换行而是end的值
        num2 += 1
    print()
    num1 += 1
```

```python
side = 1
length = 10
while side <= length:
    j = length - side
    while j >= 1:
        print(' ', end = ' ')
        j -= 1
    i = 1
    while i <= 2 * side - 1:
        print('*', end = ' ')
        i += 1
    print()
    side += 1
```

#### Break and continue
* break: 跳出循环
* continue: 跳过当次循环

```python
summation = 0
for i in range(1, 51):
    if summation >= 100:
        break
    else:
        summation += i
print(summation)
```

```python
for i in range(1, 51):
    if i % 2 == 0:
        continue
        # print('This line will not be printed')
    else:
        print(i, end = ' ')
```

```python
for i in 'I love python':
    if i == 'o':
        # break
        continue
    print(i, end = '')
```

while：未知循环次数\
for：已知循环次数

```python
for i in range(1, 10):
    for j in range(1, 11 - i):
        print('%d * %d = %d' %(i, j, j * j), end = ' ')
    print()
```

#### for - else / while - else：
只有在 break 后不执行 else 中的代码

```python
password = '123'
for i in range(1, 3):
    passwordInput = input('Enter your password: ')
    if password == passwordInput:
        print('Login successful')
        break   # 跳出for循环，不执行第9行的else
    else:
        print('Wrong password')
else:
    print('Account locked')
```

```python
index = 1
age = 25

while index <= 3:
    guess = int(input('Enter your guess: '))
    if guess == age:
        print('Congratulations, you guessed it!')
        break
    elif guess > age:
        print('The guess is too large')
    else:
        print('The guess is too small')
    index += 1

    if index == 4:
        # Input validity detection
        while True:
            answer = input('Keep going? (y/n): ')
            if answer in ['y', 'Y', 'n', 'N']:
                break
            else:
                print('Invalid input')

        if answer == 'y' or answer == 'Y':
            index = 1
        else:
            print('Thank you for playing!')
            break
```

Alternative solution:

```python
index = 1
age = 25
answer = 'y'
break_all = False

while True:
    if answer == 'y' or answer == 'Y':
        # 进行3次猜测尝试
        for i in range(3):
            guess = int(input('Enter your guess: '))
            if guess == age:
                print('Congratulations, you guessed it!')
                break_all = True
                break
            elif guess > age:
                print('Your guess is too large.')
            else:
                print('Your guess is too small.')

        if break_all:
            break

        # 验证用户输入
        while True:
            answer = input('Keep going? (y/n): ')
            if answer in ['y', 'Y', 'n', 'N']:
                break
            else:
                print('Invalid input! Please enter "y" or "n".')

    elif answer == 'n' or answer == 'N':
        print('Thank you for playing!')
        break
```
