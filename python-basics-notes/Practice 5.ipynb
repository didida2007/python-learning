```python
def cum_sum(start, end):
    return (end + start) * (end - start + 1) / 2

print(cum_sum(1, 10))
print(cum_sum(20, 30))
print(cum_sum(35, 45))
```

```python
def monk_compute(monks, chicken):
    for i in range(monks + 1):
        ate = i * 3 + (monks - i) / 3
        if ate == chicken:
            return i, monks - i
    return None

print(monk_compute(100, 100))
```

```python
def unique_num(list1):
    for item in list1:
        if list1.count(item) == 1:
            return item
    return None

print(unique_num([1, 1, 2, 3, 6, 2, 6]))
```

```python
def unique_num_set(list1):
    set1 = set(list1)
    for item in set1:
        list1.remove(item)
    set2 = set(list1)
    return set1 - set2

print(unique_num_set([1, 1, 2, 3, 6, 2, 6]))
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
scores = [85, 45, 92, 78, 58]

zipped = zip(names, scores)
filtered_zip = filter(lambda student: student[1] >= 60, list(zipped))
sorted_zip = sorted(list(filtered_zip), key=lambda student: student[1], reverse=True)
print(sorted_zip)
```

```python
string_lists = [
    ['apple', 'banana2', 'cherry'],
    ['dog', 'elephant', 'fish'],
    ['123', '456', '789']
]

num_list, over_five, has_vowel = [], [], []

for i, str_list in enumerate(string_lists):
    list_has_vowel = []

    # Find all strings with numbers
    for j, string in enumerate(str_list):
        if string.isalnum() and not string.isalpha():
            num_tuple = string, j, i
            num_list.append(num_tuple)

        # Check if all strings have vowels
        has_vowel_map = map(lambda char: char in 'aeiouAEIOU', string)
        list_has_vowel.append(any(list(has_vowel_map)))
    has_vowel.append(all(list_has_vowel))

    # Check if each list contains at least one string with length >= 5
    over_five_map = map(lambda string: len(string) >= 5, str_list)
    over_five.append(any(list(over_five_map)))

print(num_list)
print(over_five)
print(has_vowel)
```

```python
list1 = [1, 2, 3, 4, 5, 5, 4]
list2 = [4, 5, 6, 7, 8, 8]
set1, set2 = set(list1), set(list2)

print(set1 & set2)
print(set1 - set2)
print(set2 - set1)
print(set1 | set2)
```

```python
def is_anagram(string1, string2):
    return set(string1) == set(string2) and len(string1) == len(string2)

def unique_words(string):
    string_list = string.split()
    return set(string_list)

def remove_duplicates(listA):
    return set(listA)

print(is_anagram('silent', 'listen'))
print(unique_words('hello world hello python'))
print(remove_duplicates([1, 2, 2, 3, 4, 4, 5]))
```
