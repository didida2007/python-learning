```python
text = input('Please enter text:')
countL, countN, countS, countO = 0, 0, 0, 0

for char in text:
    if char.isalpha():
        countL += 1
    elif char.isdigit():
        countN += 1
    elif char.isspace():
        countS += 1
    else:
        countO += 1

print('Number of letters:', countL)
print('Number of numbers:', countN)
print('Number of spaces', countS)
print('Number of special characters:', countO)
```

```python
text = input('Please enter text:').lower()
newText = ''

for char in text:
    if char.isalnum():
        newText += char

for i in range(0, len(newText)):
    if newText[i] == newText[-1 - i]:
        pass
    else:
        print('Not a palindrome')
        break
else:
    print('This is a palindrome')
```

```python
text = input('Please enter text:')
ans = input('Uppercase (u). Lowercase (l). Title (t). Replace (r).')

if ans == 'u':
    print(text.upper())
elif ans == 'l':
    print(text.lower())
elif ans == 't':
    print(text.title())
elif ans == 'r':
    oldWord = input('Please enter replaced text')
    newWord = input('Please enter replacing text')
    print(text.replace(oldWord, newWord))
else:
    print('Invalid input')
```

```python
studentList = []
scoreList = []
ageList = []
studentNum = 0
loopCount = 0

while True:
    if loopCount > 0:
        keep = input('Keep entering? (y/n) ')
        if keep != 'y':
            print('Thank you for using')
            break
    loopCount += 1

    req = input("Add student? (1)\n"
                "Show all in descending score? (2)\n"
                "Show all in ascending age? (3)\n"
                "Find student?(4)\n"
                "Compute average score? (5)\n")

    if req == '1':
        name = input('Please enter student name: ')
        while True:
            age = input('Please enter student age: ')
            try:
                age = int(age)
                break
            except ValueError:
                print('Invalid input')
        while True:
            score = input('Please enter student score: ')
            try:
                score = int(score)
                break
            except ValueError:
                print('Invalid input')
        student = [name, age, score]
        ageList.append(age)
        scoreList.append(score)
        studentNum += 1
        studentList.append(student)

    elif req == '2' or req == '3':
        ascScoreList = scoreList[:]
        ascScoreList.sort()
        ascAgeList = ageList[:]
        ascAgeList.sort()
        indexList = []
        sortedStudentList = list(range(studentNum))

        # Find the ranking of scores and ages
        if req == '2':
            for i in range(0, studentNum):
                scoreRank = ascScoreList.index(scoreList[i])
                if scoreRank not in indexList:
                    indexList.append(scoreRank)
                else:
                    indexList.append(scoreRank + 1)
        else:
            for i in range(0, studentNum):
                ageRank = ascAgeList.index(ageList[i])
                if ageRank not in indexList:
                    indexList.append(ageRank)
                else:
                    indexList.append(ageRank + 1)
                    # 对3个相同成绩不适用，改为sort(key) + enumerate()

        # newList[ranking] = oldList
        for i in indexList:
            sortedStudentList[indexList[i]] = studentList[i]
        if req == '2':
            sortedStudentList.reverse()
        print(sortedStudentList)

    elif req == '4':
        studentName = input('Please enter student name: ')
        for student in studentList:
            if student[0] == studentName:
                print(student)
                break
        else:
            print('Student not found')

    elif req == '5':
        if studentNum == 0:
            print('No students found')
            continue
        scoreSum = 0
        for student in studentList:
            scoreSum += student[2]
        print(f'Average score: {scoreSum / studentNum}')

    else:
        print('Invalid input')
```

```python
totalItems = ()
indexId = []

while True:
    req = input("Add item? (1)\n"
                "Order by price? (2)\n"
                "Find item with price range? (3)\n"
                "Exit? (4)\n")

    if req == '1':
        itemStr = input("Please enter item info (id name price inventory)")

        # converting id, price, and inventory to num
        try:
            itemList = itemStr.split(' ')
            itemId = int(itemList[0])
            itemPrice = float(itemList[2])
            itemInventory = int(itemList[3])
        except (ValueError, IndexError):
            print('Invalid input')
            continue
        itemList[0] = itemId
        itemList[2] = itemPrice
        itemList[3] = itemInventory

        # Deal with repeating ids
        if itemId not in indexId:
            indexId.append(itemId)
        else:
            print('Repeating id')
            continue

        # Converting everything to tuples
        itemTuple = ()
        for info in itemList:
            itemTuple += tuple([info])
        totalItems +=  tuple([itemTuple])

    elif req == '2':
        totalList = list(totalItems)
        totalList.sort(key=lambda x: x[2])
        print(totalList)

    elif req == '3':
        priceStr = input("Please enter the price range (lower upper)")
        priceRange = priceStr.split(' ')

        try:
            lower = float(priceRange[0])
            upper = float(priceRange[1])
        except ValueError:
            print('Invalid input')
            continue

        count = 0
        for item in totalItems:
            if lower <= item[2] <= upper:
                print(f'ID: {item[0]}   Name: {item[1]}   Price: {item[2]}   Inventory: {item[3]}')
                count += 1
        if count == 0:
            print('No item found')

    elif req == '4':
        print('Thank you for using')
        break

    else:
        print('Invalid input')
```

```python
contacts = {}
while True:
    req = input('Add contact? (1)\n'
                'Find contact (2) \n'
                'Show all contacts? (3)\n'
                'Exit? (4)\n')

    if req == '1':
        while True:
            name = input('Please enter contact name: ')
            if name in contacts.keys():
            # if name in contacts:
                print('Name already exists')
            else:
                number = input('Please enter contact number: ')
                contacts[name] = number
                break

    elif req == '2':
        targetName = input('Please enter contact name: ')
        for name, number in contacts.items():
        # if targetName in contacts:
            if targetName == name:
                print(f'Name: {name}   Number: {number}')
                break
        else:
            print('Name not found')

    elif req == '3':
        for name, number in contacts.items():
            print(f'Name: {name}   Number: {number}')

    elif req == '4':
        print('Thank you for using')
        break

    else:
        print('Invalid input')
```

```python
studentDict = {}
while True:
    req = input('Add student? (1)\n'
                'Find student (2) \n'
                'Show all students? (3)\n'
                'Exit? (4)\n')

    if req == '1':
        while True:
            number = input('Please enter student number: ')
            if number in studentDict:
                print('Number already exists')
            else:
                name = input('Please enter student name: ')
                age = input('Please enter student age: ')
                score = input('Please enter student score: ')
                studentDict[number] = {'name': name, 'age': age, 'score': score}
                break

    elif req == '2':
        targetNum = input('Please enter student number: ')
        if targetNum in studentDict:
            info = studentDict[targetNum]
            name, age, score = info.values()
            print(f'Number: {targetNum}   Name: {name}   Age: {age}   Score: {score}')
            break
        else:
            print('Number not found')

    elif req == '3':
        for num, info in studentDict.items():
            name, age, score = info.values()
            print(f'Number: {num}   Name: {name}   Age: {age}   Score: {score}')

    elif req == '4':
        print('Thank you for using')
        break

    else:
        print('Invalid input')
```
