```python
while True:
    ans = input('Please enter a number. Enter "exit" to quit. ')
    if ans == 'exit':
        print('Thank you for playing')
        break
    else:
        try:
            num = int(ans)
            if num > 0:
                print('Positive')
            elif num == 0:
                print('Zero')
            else:
                print('Negative')
        except ValueError:
            print('Invalid input')
```

```python
for i in range(1, 101):
    count = 0
    for j in range(1, i + 1):
        if i%j == 0:
            count += 1
    if count == 2:
        print(i, end = ' ')
```

Optimization:

```python
print("1-100之间的质数有：")
# 单独处理2
print(2, end=' ')
# 只检查奇数
for i in range(3, 101, 2):
    # 检查从2到i的平方根（取整+1）之间的所有奇数
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            break
    else:
        # 如果内层循环正常结束（没遇到break），说明是质数
        print(i, end=' ')
```

```python
rows = int(input('Please enter the number of rows: '))
for i in range(1, rows + 1):
    for j in range(1, rows - i + 1):
        print(' ', end = '')
    for j in range(1, 2 * i):
        print('*', end = '')
    print()
```

Optimization:

```python
rows = int(input('Please enter the number of rows: '))
for i in range(1, rows + 1):
    # " " 重复 (rows - i) 次
    # "*" 重复 (2*i - 1) 次
    print(' ' * (rows - i) + '*' * (2 * i - 1))
```
