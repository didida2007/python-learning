## 字符串

#### 序列
* 按顺序排列的值
* 字符串，列表，元组
* 支持索引与切片
* 索引大于等于0：从左往右；索引小于0：从右往左

#### 方法

```python
test = 'i LOVE python'
print(test[0])
print(test[-1])
for letter in test:     # 字符串也是可遍历的序列
    print(letter, end = '')
```

```python
# 首字母大写，其他字母小写
print(test.capitalize())
# 整体大小写
print(test.upper())
print(test.lower())
# 单词首字母大写
print(test.title())
# 大小写切换
print(test.swapcase())
```

```python
a = '           abc            '
print(a)
# 去左右空格
print(a.strip())
# 去左/右空格
print(a.lstrip())
print(a.rstrip())
# 去左右字符
b = '~~def~~'
print(b.strip('~'))
```

```python
# 复制字符串
b = a    # 将a对象的内存地址赋给b
print('a的内存地址：', id(a))    # id 函数查看对象的内存地址
print('b的内存地址：', id(b))
```

```python
# 查找目标对象在字符串中的索引值
print(test.find('p'))
print(test.find('m'))    # 没找到则输出-1
print(test.index('p'))    # 与find()功能类似，但没找到会报错
```

```python
# 字符串是否以目标对象为起始
print(test.startswith('i'))
print(test.startswith('I'))    # 检测大小写
print(test.endswith('n'))
```

```python
# 切片
# [start : end : step]
# 默认 step = 1
print(test[2 : 6])    # [start, end)
print(test[2 :])
print(test[: 3])
print(test[:: -1])    # step < 0 为从右向左检索
```

```python
b = '23leTTer35'
# 检测字符串是否只包含字母与数字
print(b.isalnum())
# 检测字符串是否只包含字母/数字
print(b.isalpha())
print(b.isdigit())
# 检测字符串是否只包含大写/小写
print(b.isupper())
print(b.islower())
```

```python
# 将目标对象以字符串相连
c = ','
print(c.join('abc'))
```

```python
# 将字符串以分隔符分割
d = 'essence of life'
print(d.split())
print(d.split('e'))
print(d.split('e', 2))
```

```python
d = 'essence of life'
# 将字符串以（第一个）分隔符分割为三段
print(d.partition('e'))

# 替换目标对象
# replace(old, new, count = None)
print(d.replace('e', 'o', count = 2))    # count为替换数量，默认为0

# 查找目标对象数量
print(d.count('e'))
```

```python
# 返回字符串长度
print(len(test))
# 字符串是否包含目标对象
print('LOVE' in test)
```

```python
# 字符串的修改
text = 'hello'
newText = text[0] + 'a' + text[2 :]
print(newText)
```

## 列表

#### 特点
* 语法格式：[ , ]
* 数据项可以改变，内存地址不会改变
* 支持索引与切片
* 和字符串的区别：字符串不可变，而列表可变
    * 每一次字符串修改都创建了一个新字符串，因此需要赋值给自身
    * 列表修改作用于原列表，不能赋值给自身/

#### 方法

```python
items = [1, 2, 3, 'hello', True]
print(type(items))
# 返回列表的长度
print(len(items))
```

```python
print(items[0])
print(items[2 : 4])
print(items[3 :])
print(items[-2 : -4 : -1])
print(items * 2)
```

```python
# 追加
items.append(['dfg', 35])    # 将整个列表作为一个元素追加（展开添加用extend()）
items.append('appended data')
print(items)
```

```python
# 插入
items.insert(1, 'inserted data')
print(items)
```

```python
# 转换为列表
nums = list(range(6))
text = list('hello')
print(nums)
print(text)
```

```python
# 批量添加
items.extend(nums)
print(items)
```

```python
# 修改
items[2 : 4] = [3, 4]
print(items)
```

```python
# 删除
del items[3]
print(items)
```

```python
# 移除第一个指定元素
items.remove(3)
print(items)
```

```python
# 移除置顶项的索引值
items.pop()    # 移除最后一项
items.pop(3)
print(items)
```

```python
# pop和del的区别
nums1 = list(range(6))
print(nums1)
# pop删除元素并使用它的值
lastNum = nums1.pop()
print(lastNum)
print(nums1)
# del不返回任何值
del nums1[0]
print(nums1)
```

```python
# 查找目标对象在列表中的索引值
# index(element, start, end)
# 默认为 index(element, 0)
print(items.index('hello'))
# 和find()的区别：查找不到时find()返回-1，而index()报错
```

```python
# 排序（正序）
nums2 = [2, 5, 4.3]
nums2.sort()
print(nums2)

# 返回新列表
nums3 = sorted(nums2)
print(nums3)

# 反转
nums2.reverse()
print(nums2)
```

```python
# 以列表实现字符串修改
text = 'hello'
textList = list(text)
textList[1] = 'a'
newText = ''.join(textList)
print(newText)
```

## 元组

#### 特点
* 不可变的序列
* 格式：( , )
* 当元组只有一个元素时要加逗号，不然会被解释器当成整形
* 可以是任何类型

#### 方法

```python
tupleA = ()
print(type(tupleA))
tupleB = (1)
print(type(tupleB))
tupleC = (1,)
print(type(tupleC))
```

```python
tupleD = (1, 2.3, 'hello', True, [1, 2, 3], {'a', 'b'})
print(tupleD)
print(tupleD[2:4])
print(tupleD[::-2])
```

```python
# tupleD[1] = 0
# TypeError: 'tuple' object does not support item assignment
tupleD[4][1] = 0
# 元组中的列表可以修改
print(tupleD)
```

```python
tupleE = tuple(range(10))
print(tupleE)
print(tupleE.count(8))
```

```python
# 元组解包
tupleF = ('Alice', 17, 99)
name, age, score = tupleF
print(name)

# 解包与循环
tupleG = (('Alice', 17, 99), ('Bob', 18, 70), ('Cat', 19, 60))
for name, age, score in tupleG:
    print(f'{name} is {age} years old')

# 另一种解包法
tupleH = tuple(range(1, 6))
first, *middle, last = tupleH
print(first)
print(middle)
print(last)
```

```python
tupleH = (1, 2)
tupleI = (1, 2, 3)
combined = tupleH * 2 + tupleI
print(combined)
print(id(combined))

# 创建了新元组而不是在原元组上修改
combined += (1,)
print(combined)
print(id(combined))
```

## 字典

#### 特点
* 键值组成的无序集合，效率高
* 不是序列类型，是python内置的高级数据类型
* 格式：{ . }
* 字典可变，但键必须是不可变的类型
* 每个键必须是唯一的，重复的键后者会覆盖前者

#### 方法

```python
dictA = {}
print(type(dictA))
dictA['name'] = 'Alice'
print(dictA)
```

```python
dictB = {'name': 'Alice', 'height': 163}
print(dictB)
dictB['personality'] = 'outgoing'
print(dictB)

print(dictB['name'])
dictB['personality'] = 'shy'
print(dictB)
```

```python
# 获取所有键
print(dictB.keys())
# 获取所有值
print(dictB.values())
# 获取所有键值对
print(dictB.items())
```

```python
for item in dictB:    # 默认遍历键
    print(item)
for item in dictB.items():    # 返回成对的元组
    print(item)
for key, value in dictB.items():
    print(f'{key.capitalize()}: {value}')
```

```python
# 字典更新
dictB.update({'height': 159})
print(dictB)
dictB.update({'rabbit': True})
print(dictB)
```

```python
# 删除键
del dictB['personality']
print(dictB)

deleted = dictB.pop('height')
print(deleted)
print(dictB)
```

```python
# 排序
print(sorted(dictB.items(), key=lambda d: d[0]))
```

```python
# 清空字典
dictC = {'a': 1, 'b': 2}
print(dictC)
dictC.clear()
print(dictC)
```

```python
# 字典推导式
dictD = {x: x ** 2 for x in range(6)}
print(dictD)
# 过滤字典
dictE = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
dictF = {k: v for k, v in dictE.items() if v % 2 == 0}
print(dictF)
```

```python
# 合并字典
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 2, 'c': 3}
# 1:
merged = dict1 | dict2
print(merged)
# 2:
merged = {**dict1, **dict2}
print(merged)
# 3:
dict1.update(dict2)
print(dict1)
```

```python
# 安全获取值
dictG = {'a': 1, 'b': 2}
print(dictG.get('c'))
print(dictG.get('c', 0))    # 查找键为'c'的值，没找到则返回0
```
