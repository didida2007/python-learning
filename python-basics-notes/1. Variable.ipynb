## 变量：

```python
a = 1
print(a)

# 多次赋值
a = 2
print(a)

# 不用定义数据类型
a = 'abc'
print(a)
```

#### 数据类型
1. 数字（num）：整数（int），浮点（float），复数（complex），布尔值（bool）
2. 字符串（str）
3. 字典（dict）
4. 元组（tuple）
5. 列表（list）

```python
a = 1
print(type(a))
a = 'abc'
print(type(a))
a = True
print(type(a))
```

```python
b = ()  # 元组
print(type(b))
c = []  # 列表
print(type(c))
d = {}  # 字典
print(type(d))
```

#### 命名规则
1. 变量名只能包含字母（区分大小写），数字，或下划线_
2. 不能以数字开头
3. Python关键字不能作为变量名

```python
name = 'A'
Name = 'B'
age = '_1'
print(name, Name, age)
```

## 基本操作符

#### 算数运算符

```python
# 同时赋值多个变量
a, b = 5, 2

print(a + b)  # 加法
print(a - b)  # 减法
print(a * b)  # 乘法
print(a ** b) # 指数
print(a % b)  # 取余
print(a / b)  # 除法，返回float
print(a // b) # 相除取整
```

#### 比较运算符

```python
a, b = 10, 5

print(a == b) # 等于
print(a != b) # 不等于
print(a > b)  # 大于
print(a < b)  # 小于
print(a >= b) # 大于等于
print(a <= b) # 小于等于
```

#### 逻辑运算符

```python
a, b, c, d = 23, 18, 10, 3

print(a > b and c < d)  # 与门
print(a > b or c < d)   # 或门
print(not a > b)        # 非门
```

优先级：
* () > not > and > or

```python
print(2 > 1 and 2 > 4   # False
      or
      3 > 2 and 9 > 6)  # True
```

#### 赋值运算符
* 算术运算符 + '='

```python
a, b, c, d = 9, 7, 3, 2

a += b    # a = a + b
print(a)
# 不能直接 print(a += b) ，因为 print() 无法输出等式

c **= d + 2    # c = c ** (d + 2)
print(c)
```

## 格式化输出与输入

#### 占位符：%

```python
name1 = 'Abby'
name2 = 'Beth'
grade = 3

# %s: str
# %d: int (decimal)
print('We are %s and %s from grade %d' %(name1, name2, grade))
```

#### 换行：\n

```python
print('I am\ngood')
```

#### f-string

```python
name = 'Alice'
print(f'I am {name}')
```

#### 输出

```python
print('Name 1: %s' %name1)
print('Name 2: %s' %name2)
print('Grade: %d' %grade)
```

```python
print('Name 1: {}' .format(name1))
print('Name 2: {}\nGrade: {}' .format(name2, grade))
```
