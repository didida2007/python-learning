```python
def sum_par(*args):
    summation = 0
    for i in args:
        summation += i
    return summation

print(sum_par(100, 30, 90))
```

```python
def odd_index(series):
    index_list = list(series[1:: 2])
    return index_list

def odd_num(series):
    num_list = [item for i, item in enumerate(series) if item % 2 == 1]
    return num_list

print(odd_index((1, 2, 3, 4, 5, 6)))
print(odd_num((1, 2, 3, 4, 5, 6)))
```

```python
def short_dict(dictionary):
    for key, value in dictionary.items():
        if len(value) > 2:
            value = value[0: 2]
        dictionary[key] = value
    return dictionary

def short_k(**kwargs):
    for key, value in kwargs.items():
        if len(value) > 2:
            value = value[0: 2]
        kwargs[key] = value
    return kwargs

print(short_dict({'a': 'ab', 'b': 'edf', 'c': [2, 3, 4, 5]}))
print(short_k(a = 'ab', b = 'edf', c = [2, 3, 4, 5]))
```

```python
def calculate_area(shape, height, width):
    if shape == 'triangle':
        return height * width / 2
    elif shape == 'rectangle':
        return height * width
    else:
        return 'Invalid shape'

print(calculate_area('triangle', 1, 3))
```

```python
def create_student_profile(name, number, age=18, major='undecided', **other_info):
    dictionary = {'name': name, 'number': number, 'age': age, 'major': major}
    return {**dictionary, **other_info}

create_student_profile('alice', 777, 17, 'literature', rabbit=True, head=0)
```

```python
def analyze_scores(score_list):
    score_mean = sum(score_list) / len(score_list)
    score_max = max(score_list)
    score_min = min(score_list)
    pass_count = len([score for score in score_list if score >= 60])
    return score_mean, score_max, score_min, pass_count

print(analyze_scores([1, 2, 3, 4, 5, 61, 62, 63, 64, 65]))
```

```python
def reverse_string(string):
    if len(string) == 0:
        return ''
    return string[-1] + reverse_string(string[: -1])

print(reverse_string('hello'))
print(reverse_string('a'))
```

```python
def flatten_list(nested_list):
    result = []
    for item in nested_list:
        if isinstance(item, list):
            result.extend(flatten_list(item))
        else:
            result.append(item)
    return result

flatten_list([1, [2, [3, 4], 5], 6])
```
